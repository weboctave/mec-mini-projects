## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

## Change to use the new json_normalize

```python
import json
from pandas import json_normalize

# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
projects['mjtheme_namecode']
```

```python
# load json as string
jsonData = json.load((open('data/world_bank_projects.json')))
```

```python
# use normalization to create tables from nested element
mjtheme = json_normalize(jsonData, 'mjtheme_namecode')
mjtheme.head()
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
mjtheme['name'].value_counts().head(10)
```

```python
# some blank there in name, use code to see
mjtheme['code'].value_counts().head(10)
```

```python
# find distinct code name pair
codeName = mjtheme[mjtheme['name']!=''].drop_duplicates()
codeName
```

```python
len(mjtheme)
```

```python
merged_df = pd.merge(mjtheme.rename(columns={"name": "old_name"}), codeName, 
                     left_on = 'code', 
                     right_on = 'code', 
                     how='left').drop(columns=['old_name'])
merged_df
```

```python
len(merged_df)
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
# including fixed blank name
merged_df['name'].value_counts().head(10)
```

## In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# look at the blank records
mjtheme[mjtheme['name']=='']
```

```python
# Look at other blank records for code 11, just to see
mjtheme[mjtheme['code']=='11'].sort_values(['code', 'name'])
```

```python
# Look at other blank records for code 6.  Sort it so blank comes first, then do backfill later
social = mjtheme[mjtheme['code']=='6'].sort_values(['code', 'name'])
social
```

```python
socialNan = social.replace({'': None})        # replace blank with None
socialBfill = socialNan.bfill(axis ='rows')   # backfill the None
socialBfill
```

```python
# created the dataframe with name populated
mjthemeNan = mjtheme.sort_values(['code', 'name']).replace({'': None}) # blank come first then in sort   # replace blank with None
mjthemeBfill = mjthemeNan.bfill(axis ='rows')   # backfill the None with the code has name
mjthemeBfill
```

```python
mjthemeBfill[mjthemeBfill['name']=='']
```
